```python
import pandas as pd

from IPython.display import Image, display, HTML
from ipywidgets import widgets, HBox
import pandas as pd

from matplotlib.pyplot import figure, imshow, axis
from matplotlib.image import imread

def showImagesHorizontally(list_of_files):
    fig = figure(figsize=(15,15))
    number_of_files = len(list_of_files)
    for i, file in enumerate(list_of_files):
        a=fig.add_subplot(1,number_of_files,i+1)
        image = imread(file)
        imshow(image)
        axis('off')        

def make_pretty(styler, limits=[0, 1],fmt="{:.3f}", cmap="RdYlGn"):
    styler.format(fmt)
    styler.background_gradient(axis=None, vmin=limits[0], vmax=limits[1], cmap=cmap)
    return styler

def make_smaller():
    heading_properties = [('font-weitgh', 'bold'), ('font-size', '9px') ]
    cell_properties = [('font-size', '8.5px')]

    return [dict(selector="th", props=heading_properties), dict(selector="td", props=cell_properties)]
```

```python
columns_drop = ["precision", "recall", "support"]

df_text_re = pd.read_csv("../../src/models/text/text_neural_simple/clf_report.csv", index_col="Unnamed: 0")
df_image_mob_crop = pd.read_csv("../../src/models/image/image_mobilenet_224_crop/clf_report.csv", index_col="Unnamed: 0")
df_fusion = pd.read_csv("../../src/models/fusion/fusion_mobilenet_simple_224/clf_report.csv", index_col="Unnamed: 0")

df_text_re.drop(columns_drop, axis=1, inplace=True)
df_image_mob_crop.drop(columns_drop, axis=1, inplace=True)
df_fusion.drop(columns_drop, axis=1, inplace=True)

df_text_re.rename({"f1-score": "Text f1-score"}, axis=1, inplace=True)
df_image_mob_crop.rename({"f1-score": "Image f1-score"}, axis=1, inplace=True)
df_fusion.rename({"f1-score": "Fusion f1-score"}, axis=1, inplace=True)

df = pd.concat([df_text_re, df_image_mob_crop, df_fusion], axis=1)
#df.columns = [column[5:] for column in df.columns]
df.drop(["accuracy", "macro avg"], axis=0, inplace=True)
#
df.style.pipe(make_pretty).set_table_styles(make_smaller())
```
